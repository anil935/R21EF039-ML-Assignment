```python
import pandas as pd
```

```python
df=pd.read_csv("C:/Users/Vsrvr/Desktop/stroke.csv")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.hist(df['age'], bins=20, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(6, 6))
sns.countplot(x='gender', data=df, palette='pastel')
plt.title('Gender Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
sns.countplot(x='hypertension', data=df, palette='Set2')
plt.title('Hypertension Distribution')

plt.subplot(1, 2, 2)
sns.countplot(x='heart_disease', data=df, palette='Set2')
plt.title('Heart Disease Distribution')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.countplot(x='ever_married', data=df, palette='muted')
plt.title('Marriage Status')

plt.subplot(1, 2, 2)
sns.countplot(x='work_type', data=df, palette='muted')
plt.title('Work Type')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.countplot(x='Residence_type', data=df, palette='dark')
plt.title('Residence Type')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['avg_glucose_level'], kde=True, color='salmon')
plt.title('Average Glucose Level Distribution')
plt.xlabel('Average Glucose Level')
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['bmi'], kde=True, color='green')
plt.title('BMI Distribution')
plt.xlabel('BMI')
plt.ylabel('Density')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='smoking_status', data=df, palette='bright')
plt.title('Smoking Status Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6, 6))
sns.countplot(x='stroke', data=df, palette='colorblind')
plt.title('Stroke Distribution')
plt.show()
```
